```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("./Motor_Vehicle_Collisions_-_Crashes.csv")
```

**Univariate** **Analysis**

# **Histogram**


```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df["CRASH DATE"], bins=10)
plt.xlabel("date of crash")
plt.ylabel("Number of crashes")
plt.title("Histogram of car_crashes")
plt.show()
```

Inference for the above histogram--
The above histogram is plottted between Number of crashes and date of crash where Number if crashes is plotted on y-axis and data of crash is plotted on x-axis and the histogram depicts that the number of crashes are huge for every day but it was controlled in a constructive way such that the number of crashes became less and and the again due to some rules not followed by the people the crashes increased day by day.

**Boxplot**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="NUMBER OF PERSONS INJURED", data=df)
plt.xlabel("Number_of_persons_injured")
plt.ylabel("Value")
plt.title("Boxplot of Number_of_people_injured")
plt.show()
```

## **Pie Chart**

```python
plt.figure(figsize=(20, 10))
plt.pie(df["CONTRIBUTING FACTOR VEHICLE 1"].value_counts().sort_values(ascending=False).values, labels=df["CONTRIBUTING FACTOR VEHICLE 1"].value_counts().sort_values(ascending=False).index, autopct="%1.1f%%")
plt.title("Pie chart of contributing factor vehicle 1")
plt.show()
```

**Bar Chart**

```python
plt.figure(figsize=(10, 6))
plt.bar(df["NUMBER OF PEDESTRIANS KILLED"].value_counts().index, df["NUMBER OF PEDESTRIANS KILLED"].value_counts().values)
plt.xlabel("number_of_pedestrians_killed")
plt.ylabel("Number_of_pedestrians_killed")
plt.title("Bar chart of number_of_persons_killed")
plt.show()
```

Bi-Variate Analysis


**Scatter Plot**

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["COLLISION_ID"], df["COLLISION_ID"])
plt.xlabel("collision_id")
plt.ylabel("number_of_collisions")
plt.title("Scatter plot of collision_id vs. number_of_collisions")
plt.show()
```

**Line Chart**

```python
plt.figure(figsize=(10, 6))
plt.plot(df["CRASH DATE"], df["LATITUDE"])
plt.xlabel("CRASH DATE")
plt.ylabel("CROSS STREET NAME")
plt.title("Line chart of CRASH DATE vs. CRASH TIME")
plt.show()
```

**Joint Plot**

```python
sns.jointplot(x="NUMBER OF PERSONS KILLED", data=df)
plt.title("Joint plot of number_of_persons_killed vs. Date")
plt.show()
```
